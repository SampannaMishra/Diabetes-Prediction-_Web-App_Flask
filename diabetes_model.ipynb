```python
df = data.copy()
```

```python
cols = ['BloodPressure', 'Glucose','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age'] 
```

```python
# removing outliers from the dataset
cols = ['BloodPressure', 'Glucose','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age'] # one or more

Q1 = df[cols].quantile(0.25)
Q3 = df[cols].quantile(0.75)
IQR = Q3 - Q1

df = df[~((df[cols] < (Q1 - 1.5 * IQR)) |(df[cols] > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
df['SkinThickness'] = df['SkinThickness'].replace(0,df['SkinThickness'].median())
```

```python
df['Insulin'] = df['Insulin'].replace(0,df['Insulin'].median())
```

```python
df.head()
```

```python
# dependent and independent variables

X = df.iloc[:,0:8].values
Y = df.iloc[:, 8].values
```

```python
# spliting the dataset
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2,random_state=0)
```

```python
# standardization on train and test data
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# fitting linear regression model
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train,Y_train)
```
